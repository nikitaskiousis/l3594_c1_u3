```python
import pandas as pd
import ppscore as pps
import seaborn as sns
```

```python
df = pd.read_csv("https://media.githubusercontent.com/media/kdemertzis/EKPA/main/Data/Ransomware.csv",sep='|')
df
```

```python
irrelevant_columns = ['Name', 'md5', 'Machine', 'SizeOfOptionalHeader', 'Characteristics']
df_relevant = df.drop(columns=irrelevant_columns, errors='ignore')
df_relevant
```

```python
pps_matrix = pps.matrix(df_relevant)
```
